```python
consumo_diario = [50, 45, 55, 60, 48, 52, 47, 49, 53, 51,
                  50, 46, 54, 61, 49, 53, 48, 50, 52, 47,
                  56, 50, 55, 60, 48, 52, 47, 49, 53, 51]
```

```python
def consumo(lista):
    soma = sum(lista)
    dias = len(lista)
    media = soma / dias
    return media
```

```python
print(f'O consumo diário médio foi: {consumo(consumo_diario):.2f} kWh')
print(f'O consumo diário médio foi: {consumo(consumo_diario)} kWh')
```
